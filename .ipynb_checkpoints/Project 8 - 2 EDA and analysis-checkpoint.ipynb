```python
# encoding for categorical variables with levels > 2
lt1_dummies = pd.get_dummies(lt1, drop_first=True, prefix_sep='.', columns=['Age'], dtype=int)
#lt1_dummies
```

```python
# spliting for test and train
Y = lt1_dummies[['Stroke']]
```

```python
X = lt1_dummies.loc[:, lt1_dummies.columns != 'Stroke']
```

```python
# using a logistic regression modeling example found here:
# https://towardsdatascience.com/machine-learning-projects-on-the-cloud-key-steps-in-the-process-d61f67d9d454

oversample = SMOTE()
os = SMOTE(random_state=0)
```

```python
# using smote to balance the sample
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

columns = X_train.columns
```

```python
os_data_X,os_data_y=os.fit_resample(X_train, y_train)
```

```python
os_data_X = pd.DataFrame(data=os_data_X,columns=columns )
os_data_y= pd.DataFrame(data=os_data_y,columns=['Stroke'])
```

```python
# SMOTE balanced the data - reviewing the change
print("length of oversampled data is ",len(os_data_X))
print("Number of Yes Stroke in oversampled data",len(os_data_y[os_data_y['Stroke']==0]))
print("Number of No Stroke",len(os_data_y[os_data_y['Stroke']==1]))
print("Proportion of Yes Stroke data in oversampled data is ",len(os_data_y[os_data_y['Stroke']==0])/len(os_data_X))
print("Proportion of No Stroke data in oversampled data is ",len(os_data_y[os_data_y['Stroke']==1])/len(os_data_X))
```

```python
# feature selection, seeing if features can be dropped
logreg = LogisticRegression(max_iter=1000)
```

```python
#Recursive Feature Elimination (RFE) is based on the idea to repeatedly construct a model and choose either the best or worst performing feature,
#setting the feature aside and then repeating the process with the rest of the features

rfe = RFE(logreg, n_features_to_select=20)
```

```python
rfe = rfe.fit(os_data_X, os_data_y.values.ravel())
print(rfe.support_)
print(rfe.ranking_)
```

```python
Xy = os_data_X.columns[rfe.support_]
print("Selected features:", Xy)
```

```python
Xn = os_data_X[['HighChol', 'CholCheck', 'Smoker', 'HeartDiseaseorAttack',
       'HvyAlcoholConsump', 'DiffWalk', 'HighBP', 'Diabetes', 'Age.2.0',
       'Age.3.0', 'Age.4.0', 'Age.5.0', 'Age.6.0', 'Age.7.0', 'Age.8.0',
       'Age.9.0', 'Age.10.0', 'Age.11.0', 'Age.12.0', 'Age.13.0']]
```

```python
Yn = os_data_y['Stroke']
```

```python
# initial modeling
# checking the p value to see if other columns should come out with pvalues higher than .05 but none are found

import statsmodels.api as sm
logit_model=sm.Logit(Yn.astype(float),Xn.astype(float))
result=logit_model.fit()
print(result.summary2())
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xn, Yn, test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
print('Confusion Matrix')
ConfusionMatrixDisplay.from_predictions(y_test, y_pred);
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
y_test_int = y_test
auc_lr = roc_auc_score(y_test_int, y_pred)
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test_int, y_pred)
```

```python
y_test_int = y_test
auc_lr = roc_auc_score(y_test_int, y_pred)
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test_int, y_pred)

plt.plot(fpr_lr, tpr_lr, label=f'AUC (Logistic Regression) = {auc_lr:.2f}')
plt.plot([0, 1], [0, 1], color='blue', linestyle='--', label='Baseline')
plt.title('Stroke - ROC Curve', size=12)
plt.xlabel('False Positive Rate', size=12)
plt.ylabel('True Positive Rate', size=12)
plt.legend();
```
